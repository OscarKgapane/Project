```python
class FIFAPlayer:

    def __init__(self, name, position, rating = 60):
      self.name = name
      self.position = position
      self.rating = rating


    def player_info(self):
      print(f"Name: {self.name}")
      print(f"Postion: {self.position}")
      print(f"Player Rating: {self.rating}")


    def player_goal(self):
      print(f"{self.name} has scored a goal!")

    def improve_rating(self, points):


      self.rating += points

      print(f"{self.name}'s rating has improved by {points} to {self.rating}")


player1 = FIFAPlayer("Lionel Messi", "Rightwinger", 80 )
player1.player_info()
player1.player_goal()
player1.improve_rating(3)
print("\n")
player1.player_info()
print("\n")

player2 = FIFAPlayer("David Beckham", "Midfielder", 89 )
player2.player_info()
player2.player_goal()
player1.player_goal()
player2.improve_rating(5)
print("\n")
player2.player_info()
print("/n")

player3 = FIFAPlayer("Christiano Ronaldo", "Forward", 94)
player3.player_info()
player3.player_goal()
player3.improve_rating(1)
print("\n")

player_list = []
player_list.append(player1)
player_list.append(player2)
player_list.append(player3)



for FIFAPlayer in player_list:
  FIFAPlayer.player_info()
  print("\n")


my_list = [1, 2, 3, 4]
print(my_list)
my_list.append(5)
print(my_list)


```
