```python
class Pokemon:
    def __init__(self, name, p_type, level =1):
        self.name = name
        self.p_type = p_type
        self.level = level

    def display_info(self):
        print(f"Name: {self.name}")
        print(f"Type: {self.p_type}")
        print(f"Level: {self.level}")


    def attack(self):
        print(f"{self.name} attack with a level {self.level} attack")

    def level_up(self):
        self.level += 1
        print(f"{self.name} leveled up to level {self.level}!")

pikachu = Pokemon("Pikachu", "Electric")

pikachu.display_info()
pikachu.level_up()
pikachu.attack()

pikachu.display_info()
print("\n")

bulbasaur = Pokemon("Bulbasaur", "Grass/Poison")
bulbasaur.display_info()
bulbasaur.level_up()
bulbasaur.level_up()
bulbasaur.attack()
bulbasaur.display_info()

print("\n")
charmander = Pokemon("Charmander", "Fire",)
charmander.display_info()
charmander.level_up()
charmander.level_up()
charmander.level_up()
charmander.attack()
charmander.display_info()

print("\n")
squirtle = Pokemon("Squirtle", "Water",)
squirtle.display_info()
squirtle.level_up()
squirtle.attack()
squirtle.display_info()

print("\n")

pokemon_list = [pikachu, bulbasaur, charmander, squirtle]
pokedex_welcome = "Welcome: Your Pokedex Contains"

print("*" * len(pokedex_welcome))
print(pokedex_welcome)
print("*" * len(pokedex_welcome))


for pokemon in pokemon_list:
  pokemon.display_info()

  print("\n")











```
